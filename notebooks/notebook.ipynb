```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
df_raw = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/regularized-linear-regression-project-tutorial/main/demographic_health_data.csv')
```

```python
df_raw.sample(10)
```

```python
df_raw.to_csv('../data/raw/data_raw.csv')
```

```python
df_processed = df_raw.copy()
```

```python
df_interin = df_processed[["0-9","19-Oct","20-29","30-39","40-49","50-59","60-69","70-79","80+","% White-alone","% Black-alone","% NA/AI-alone","% Asian-alone","% Hawaiian/PI-alone","Less than a high school diploma 2014-18","High school diploma only 2014-18","Bachelor's degree or higher 2014-18","MEDHHINC_2018","Employed_2018","Unemployed_2018","Active Physicians per 100000 Population 2018 (AAMC)","Obesity_prevalence","Heart disease_prevalence","diabetes_prevalence","CKD_prevalence","COPD_prevalence"]]
```

```python
X = df_interin.drop(['Heart disease_prevalence'],axis=1)
y = df_interin['Heart disease_prevalence']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2115)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
cv = RepeatedKFold(n_splits=10,n_repeats=3,random_state=2119)
```

```python
model = LassoCV(alphas=np.arange(0,1,0.05),cv=cv,n_jobs=-1)
```

```python
model.fit(X_train,y_train)
```

```python
model.alpha_
```

```python
y_hat = model.predict(X_test)
```

```python
print(f'MSE: {mean_squared_error(y_hat,y_test)}')
print(f'r2: {r2_score(y_hat,y_test)}')
```
